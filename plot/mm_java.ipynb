```python
import os
import os.path as path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import FormatStrFormatter
import matplotlib.patches as mpatches
import math
import statistics as stat

# width = 505.89
width = 417.68646

inst_type = ["w/o canon", "canon"]
dims = np.array([100, 200, 500, 1000, 2000])
nan_rates = np.array([0, 1, 10, 100])
dim_labels = ["100", "200", "500", "1000", "2000"]
```

```python
def set_size(width, fraction=1):
    """ Set aesthetic figure dimensions to avoid scaling in latex.
    Parameters
    ----------
    width: float
            Width in pts
    fraction: float
            Fraction of the width which you wish the figure to occupy
    Returns
    -------
    fig_dim: tuple
            Dimensions of figure in inches
    """
    # Width of figure
    fig_width_pt = width * fraction

    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio

    fig_dim = (fig_width_in, fig_height_in)

    return fig_dim
```

```python
def pretty_print(avg_values,nan_rate):
    print(f'===== NaN rate = {nan_rate}=====')
    print("dim,nocanon,canon")
    for i in range(len(dims)):
        print(f'{dims[i]},{avg_values[0][i]},{avg_values[1][i]}')
```

```python
def plot_line(dirpath, outdir, nan_rate):
    w = 0.1
    outfile = f'nan{nan_rate}_line.pdf'
    outpath = path.join(outdir, outfile)
    colors = ['darkorchid','forestgreen','#DC7A23']
    fig, ax = plt.subplots(1,1,figsize=set_size(width, fraction=1))

    exp_cnt = len(inst_type)
    avg_values = list()
    stddev_values = list()
    min_values = list()
    max_values = list()
    for i in range(exp_cnt):
        avg_values.append(list())
        stddev_values.append(list())
        min_values.append(list())
        max_values.append(list())

    # 0: raw, 1: canon
    for d in dims:      
        input_data = pd.read_csv(path.join(dirpath, f'raw_{d}_{nan_rate}.csv'))
        avg_values[0].append(input_data["avg"].tolist()[0])
        stddev_values[0].append(input_data["stddev"].tolist()[0])
        min_values[0].append(input_data["min"].tolist()[0])
        max_values[0].append(input_data["max"].tolist()[0])        
        
        input_data = pd.read_csv(path.join(dirpath, f'canon_{d}_{nan_rate}.csv'))
        avg_values[1].append(input_data["avg"].tolist()[0])
        stddev_values[1].append(input_data["stddev"].tolist()[0])
        min_values[1].append(input_data["min"].tolist()[0])
        max_values[1].append(input_data["max"].tolist()[0])
        
    for i in range(exp_cnt):
        ax.plot(dims, avg_values[i], marker='x', color=colors[i])
#         plt.errorbar(dims, avg_values[i], yerr=stddev_values[i], marker='x', color=colors[i])    
#     for i in range(exp_cnt):
#         ax.fill_between(dims, min_values[i], max_values[i], alpha=0.4, color=colors[i])
    
    plt.xscale('log')
    plt.yscale('log')
    plt.xticks(dims, dims, fontsize=12)
    plt.yticks(fontsize=12)
    plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%g'))
    
    plt.title(f'NaN rate = {nan_rate}%')
    plt.xlabel("Matrix dimension", fontsize=12)
    plt.ylabel("Latency (s)", fontsize=12)
    
    plt.legend(inst_type, fontsize=12, title_fontsize=12, ncol=1)
    plt.grid(linestyle='--', color='silver', axis='y', linewidth=0.5)
    plt.savefig(outpath, bbox_inches='tight', format='pdf')
```

```python
def plot_bar(dirpath, outdir, nan_rate):
    w = 0.1
    outfile = f'nan{nan_rate}_bar.pdf'
    outpath = path.join(outdir, outfile)
#     colors = ['darkorchid','#DC7A23','forestgreen']
#     colors = ['#2E78D1', '#D1872E']
    colors = ['#2DABD2', '#D2542D']
    
    fig, ax = plt.subplots(1,1,figsize=set_size(width, fraction=1))

    exp_cnt = len(inst_type)
    avg_values = list()
    stddev_values = list()
    min_values = list()
    max_values = list()
    for i in range(exp_cnt):
        avg_values.append(list())
        stddev_values.append(list())
        min_values.append(list())
        max_values.append(list())

    # 0: raw, 1: canon
    for d in dims:      
        input_data = pd.read_csv(path.join(dirpath, f'raw_{d}_{nan_rate}.csv'))
        avg_values[0].append(input_data["avg"].tolist()[0])
        stddev_values[0].append(input_data["stddev"].tolist()[0])
        min_values[0].append(input_data["min"].tolist()[0])
        max_values[0].append(input_data["max"].tolist()[0])        
        
        input_data = pd.read_csv(path.join(dirpath, f'canon_{d}_{nan_rate}.csv'))
        avg_values[1].append(input_data["avg"].tolist()[0])
        stddev_values[1].append(input_data["stddev"].tolist()[0])
        min_values[1].append(input_data["min"].tolist()[0])
        max_values[1].append(input_data["max"].tolist()[0])
        
    pretty_print(avg_values,nan_rate)
    x = np.arange(len(dims))
    ax.bar(x-w/2, avg_values[0], w, color=colors[0])
    ax.bar(x+w/2, avg_values[1], w, color=colors[1])

    plt.yscale('log')
    plt.xticks(x, dims, fontsize=12)
    plt.yticks(fontsize=12)
    plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%g'))
    
    plt.title(f'NaN rate = {nan_rate}%')
    plt.xlabel("Matrix dimension", fontsize=12)
    plt.ylabel("Latency (s)", fontsize=12)
    
    plt.legend(inst_type, fontsize=12, title_fontsize=12, ncol=1)
    plt.grid(linestyle='--', color='silver', axis='y', linewidth=0.5)
    plt.savefig(outpath, bbox_inches='tight', format='pdf')
#     plt.close(fig)
```

```python
dirpath = "./data/nan_java/final"
outdir = "./figures/nan_java"
plot_bar(dirpath, outdir, 0)
plot_bar(dirpath, outdir, 1)
plot_bar(dirpath, outdir, 10)
plot_bar(dirpath, outdir, 100)
```

```python
def plot_slowdown(dirpath, outdir):
    w = 0.1
    outfile = f'slowdown.pdf'
    outpath = path.join(outdir, outfile)
    colors = ['#DFB320', '#437BBC']
    
    fig, ax = plt.subplots(1,1,figsize=set_size(width, fraction=1))
  
    input_data = pd.read_csv(path.join(dirpath, f'sd.csv'))
    java_sd = input_data["java"].tolist()
    wasm_sd = input_data["wasm"].tolist()

    x = np.arange(len(dims))
    ax.bar(x-w/2, java_sd, w, color=colors[0])
    ax.bar(x+w/2, wasm_sd, w, color=colors[1])

#     plt.yscale('log')
    plt.xticks(x, dims, fontsize=12)
    new_yts = [0.5,1,1.5]
    plt.yticks(new_yts, new_yts, fontsize=12)
    plt.ylim((0.5,1.5))
        
#     plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%g'))
    
#     plt.title(f'NaN rate = {nan_rate}%')
    plt.xlabel("Matrix dimension", fontsize=12)
    plt.ylabel("Relative NaN\ncanonicalization slowdown", fontsize=12)
    
    plt.legend(["Java", "WASM"], fontsize=11, title_fontsize=12, ncol=1)
    plt.grid(linestyle='--', color='silver', axis='y', linewidth=0.5)
    plt.savefig(outpath, bbox_inches='tight', format='pdf')
#     plt.close(fig)
```

```python
dirpath = "./data/slowdown"
outdir = "./figures/slowdown"
plot_slowdown(dirpath, outdir)
```
